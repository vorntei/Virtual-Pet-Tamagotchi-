```python
The Goal: 
Create a Pet class with attributes like name, hunger, energy, and happiness. 
Create methods like feed(), play(), and sleep() that alter these attributes.
Concepts Tested: Classes, __init__, self, instance attributes, methods.
```

```python
class Cat:
    species = "hungry"
    def __init__(self, name):
        self.name = name
cat1 = Cat("KOI")
cat2 = Cat("Tea")
cat3 = Cat("KOY")
print(cat1.name, cat1.species)
print(cat2.name, cat2.species)
print(cat3.name, cat3.species)
```

```python
class Cat:
    species = "cat"
    def __init__(self, name):
        self.name = name
        self.hunger = 100
        self.energy = 40
    def feed(self):
        self.hunger -= 20
    def sleep(self):
        self.energy +=40
cat1 = Cat("KOI")
cat1.feed()
print(cat1.name, cat1.hunger)
cat2 = Cat("Tube")
cat2.sleep()
print(cat2.name, cat2.energy)
```
